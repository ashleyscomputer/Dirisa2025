```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)

print("=== LOADING DATA ===")
df = pd.read_csv("Parquet_file.csv")

rows_before, cols_before = df.shape
print(f"Original shape: {rows_before} rows, {cols_before} columns\n")

# Overall missing + duplicates BEFORE cleaning
missing_before = df.isna().sum().sum()
dupes_before = df.duplicated().sum()

print("=== BEFORE CLEANING ===")
print(f"Total missing values (all columns): {missing_before}")
print(f"Total duplicate rows: {dupes_before}\n")

# ================================
# 2. DATETIME CONVERSION & CLEANING
# ================================
print("=== CLEANING: TIMESTAMP & DUPLICATES ===")

# Convert unitdatetime to proper datetime
df["unitdatetime"] = pd.to_datetime(df["unitdatetime"], errors="coerce")

# Drop rows with null unitdatetime
rows_before_dropna = len(df)
df = df.dropna(subset=["unitdatetime"])
rows_after_dropna = len(df)
dropped_null_ts = rows_before_dropna - rows_after_dropna

# Drop duplicate rows
df = df.drop_duplicates()
rows_after = len(df)
dropped_dupes = rows_after_dropna - rows_after

# Recompute missing + duplicates AFTER cleaning
missing_after = df.isna().sum().sum()
dupes_after = df.duplicated().sum()

print(f"Rows removed due to NULL unitdatetime: {dropped_null_ts}")
print(f"Rows removed as duplicates:         {dropped_dupes}")
print(f"Final shape after cleaning:         {rows_after} rows, {df.shape[1]} columns\n")

print("Missing values AFTER cleaning (all columns):", missing_after)
print("Duplicate rows AFTER cleaning:", dupes_after)

# Sort chronologically (important for time series)
df = df.sort_values("unitdatetime")

# ================================
# 3. FEATURE ENGINEERING (OVERSPEED)
# ================================
print("\n=== FEATURE ENGINEERING: OVERSPEED FLAG ===")

if "speed" in df.columns and "roadspeed" in df.columns:
    df["overspeed"] = df["speed"] > df["roadspeed"]
    overspeed_rate = df["overspeed"].mean() * 100
    print(f"Overspeed flag created. Overall overspeed rate: {overspeed_rate:.2f}%")
else:
    df["overspeed"] = False
    print("Columns 'speed' and/or 'roadspeed' not found. 'overspeed' set to False for all rows.")

# ================================
# 4. BASIC STRUCTURE & SUMMARY
# ================================
print("\n=== BASIC STRUCTURE ===")
print(df.info())

print("\n=== SAMPLE ROWS (HEAD) ===")
display(df.head(10))

key_numeric_cols = [c for c in ["speed", "roadspeed", "latitude", "longitude", "altitude"] if c in df.columns]

if key_numeric_cols:
    print("\n=== NUMERIC SUMMARY (KEY COLUMNS) ===")
    display(df[key_numeric_cols].describe())
else:
    print("\nNo key numeric columns found for summary.")

print("\n=== MISSING VALUES PER COLUMN ===")
print(df.isna().sum())

# ================================
# 5. EVENT DISTRIBUTION
# ================================
if "eventid" in df.columns:
    print("\n=== EVENT DISTRIBUTION ===")
    event_counts = df["eventid"].value_counts()
    display(event_counts)

    plt.figure()
    event_counts.plot(kind="bar")
    plt.title("Event Distribution")
    plt.xlabel("eventid")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
else:
    print("\nColumn 'eventid' not found. Skipping event distribution.")

# ================================
# 6. SPEED ANALYSIS
# ================================
if "speed" in df.columns:
    print("\n=== SPEED DISTRIBUTION ===")

    plt.figure()
    sns.histplot(df["speed"], kde=True, bins=40)
    plt.title("Speed Distribution")
    plt.xlabel("Speed (km/h)")
    plt.tight_layout()
    plt.show()

    if "overspeed" in df.columns:
        print("\nOverspeed value counts:")
        print(df["overspeed"].value_counts())

        plt.figure()
        df["overspeed"].value_counts().plot(kind="bar")
        plt.title("Overspeed vs Within Limit")
        plt.xticks([0, 1], ["Within limit", "Overspeed"], rotation=0)
        plt.ylabel("Count")
        plt.tight_layout()
        plt.show()
else:
    print("\nColumn 'speed' not found. Skipping speed analysis.")

# ================================
# 7. GEO SPREAD (LAT/LON)
# ================================
if "latitude" in df.columns and "longitude" in df.columns:
    print("\n=== GEO SPREAD (LAT/LON SCATTER) ===")

    plt.figure(figsize=(6, 6))
    plt.scatter(df["longitude"], df["latitude"], s=2, alpha=0.3)
    plt.title("GPS Point Scatter")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.tight_layout()
    plt.show()
else:
    print("\nLatitude/longitude columns not found. Skipping geo scatter.")

# ================================
# 8. G-FORCE ANALYSIS
# ================================
g_cols = [c for c in [
    "gforce_forward", "gforce_backward",
    "gforce_left", "gforce_right",
    "gforce_up", "gforce_down"
] if c in df.columns]

if g_cols:
    print("\n=== G-FORCE PEAKS (HARSH EVENTS) ===")

    plt.figure()
    df[g_cols].max().plot(kind="bar")
    plt.title("Peak G-Force Values")
    plt.ylabel("Max value")
    plt.tight_layout()
    plt.show()
else:
    print("\nNo g-force columns found. Skipping g-force analysis.")

# ================================
# 9. TIME SERIES ANALYSIS
# ================================
print("\n=== TIME SERIES ANALYSIS ===")

# Hour-of-day feature
df["hour"] = df["unitdatetime"].dt.hour

# Speed over time
if "speed" in df.columns:
    plt.figure(figsize=(14, 4))
    plt.plot(df["unitdatetime"], df["speed"], linewidth=0.4)
    plt.title("Speed Over Time")
    plt.xlabel("Time")
    plt.ylabel("Speed (km/h)")
    plt.tight_layout()
    plt.show()

    # Average speed by hour
    plt.figure()
    sns.lineplot(data=df, x="hour", y="speed")
    plt.title("Average Speed by Hour of Day")
    plt.xticks(range(0, 24))
    plt.xlabel("Hour of day")
    plt.ylabel("Average speed (km/h)")
    plt.tight_layout()
    plt.show()

# Set datetime index for resampling
df_ts = df.set_index("unitdatetime").sort_index()

# Event frequency over time
if "eventid" in df_ts.columns:
    events_by_hour = df_ts["eventid"].resample("1H").count()

    plt.figure(figsize=(14, 4))
    plt.plot(events_by_hour.index, events_by_hour.values)
    plt.title("Event Frequency per Hour")
    plt.xlabel("Time")
    plt.ylabel("Number of events")
    plt.tight_layout()
    plt.show()

# Overspeed rate over time
if "overspeed" in df_ts.columns:
    overspeed_ts = df_ts["overspeed"].resample("1H").mean()

    plt.figure(figsize=(14, 4))
    plt.plot(overspeed_ts.index, overspeed_ts.values)
    plt.title("Overspeed Rate Over Time")
    plt.xlabel("Time")
    plt.ylabel("Overspeed proportion")
    plt.tight_layout()
    plt.show()

print("\n=== PIPELINE COMPLETE: CLEANING + EDA + TIME SERIES DONE ===")
```

1. Overview
This dataset contains 58 426 telemetry records with 34 fields, capturing real-time vehicle behaviour such as speed, g-forces, GPS location, and event codes.
The goal is to extract meaningful safety insights and identify movement patterns over space and time.

2. Key variables:
Time: unitdatetime
GPS: latitude, longitude, altitude
Movement: speed, roadspeed, heading, bearing, odometer
Safety signals: gforce_forward, gforce_backward, gforce_left, gforce_right, gforce_up, gforce_down
Events: eventid
Location context: suburb, town, municipality, province, postal_code

3. Data Preparation:
To make the data usable, a few essential cleaning steps were applied:

Converted unitdatetime into a proper datetime format
Sorted the dataset chronologically
Removed rows missing timestamps\
Added an overspeed indicator:
df["overspeed"] = df["speed"] > df["roadspeed"]
This compares the drivers speed 

4. Event Distribution:
Event logging is heavily skewed
â€œ02â€ dominates the dataset, representing routine heartbeat/position updates

Less frequent events (â€œ00â€, â€œ03â€, â€œ5Eâ€, â€œ5Fâ€) are more meaningful for safety analysis and should be evaluated separately

5. Speed Behaviour

Key patterns:
Average speed: ~29 km/h
Typical driving: lowâ€“medium city speeds
Occasional peaks up to 121 km/h

Overspeed flag highlights when the vehicle exceeded posted limits

This gives a quick view of general driving behaviour and potential risk windows.

6. Geographic Patterns

GPS points cluster primarily around the City of Johannesburg and nearby suburbs such as Bramley and Wynberg.
These clusters help identify:
Common routes
Repeated high-risk areas
Location-specific behaviour patterns

7. G-force readings capture how the vehicle moves:

Forward/backward: braking or accelerating
Left/right: cornering
Up/down: bumps or rough terrain

Spikes in these values flag possible harsh events or poor road surfaces.

8. Time-Series Highlights
Speed Over Time: shows movement spikes and idle periods
Hourly Speed Patterns: reveals consistent daily driving rhythms
Overspeed Trend: identifies the specific hours where risky driving increases

In Conclusion:
Time-series quality is strong after cleaning

Event distribution is unbalanced but informative

Driving behaviour is mostly urban and low-speed

Overspeeding + g-force spikes are key risk markers

Spatial clustering reveals clear operational zones and potential hotspots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("dft-road-casualty-statistics-collision-provisional-2025.csv")  # replace with your file path

# Quick peek
print("First 5 rows:")
display(df.head())

# Dataset info
print("\nDataset info:")
df.info()

# Numeric summary
print("\nSummary statistics for numeric columns:")
display(df.describe())

# Dataset shape
print("\nDataset shape (rows, columns):", df.shape)
```

```python
# Accidents per year
year_counts = df['collision_year'].value_counts().sort_index()
plt.figure(figsize=(8,4))
sns.barplot(x=year_counts.index, y=year_counts.values, palette='Reds')
plt.title("Number of Collisions per Year")
plt.ylabel("Number of Collisions")
plt.xlabel("Year")
plt.show()
```

```python
severity_counts = df['collision_severity'].value_counts().sort_index()
plt.figure(figsize=(6,4))
sns.barplot(x=severity_counts.index, y=severity_counts.values, palette='coolwarm')
plt.title("Collision Severity Distribution")
plt.xlabel("Severity Level (1=Fatal, 2=Serious, 3=Slight)")
plt.ylabel("Number of Collisions")
plt.show()
```

```python
# Accidents by urban/rural area
area_counts = df['urban_or_rural_area'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=area_counts.index, y=area_counts.values, palette='viridis')
plt.title("Collisions by Urban/Rural Area")
plt.xlabel("Urban (1) / Rural (2)")
plt.ylabel("Number of Collisions")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='number_of_vehicles', data=df, palette='magma')
plt.title("Number of Vehicles Involved per Collision")
plt.xlabel("Vehicles Involved")
plt.ylabel("Count")
plt.show()
```
